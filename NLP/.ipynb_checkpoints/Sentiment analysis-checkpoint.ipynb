```python
import pandas as pd
from textblob import TextBlob
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv (r'Womens Clothing E-Commerce Reviews.csv')
```

```python
df.head()
```

```python
df.shape
```

### little bit of dataframe cleaning and organising

```python
df.columns=[x.replace(' ','_') for x in df.columns]
```

```python
df.columns = df.columns.str.lower()
```

```python
df = df[['title','review_text', 'rating', 'division_name', 'department_name']]
```

```python
df.head()
```

```python
#Getting rid of empty fields in review text
df['review_text'].replace('', np.nan, inplace=True)
df.dropna(subset=['review_text'], inplace=True)
```

```python
df.info()
```

```python
gdf = df.groupby(['division_name', 'department_name']).rating.mean()
```

```python
#Pulling out the text from review text, ready for analysis. 
blob = [TextBlob(i) for i in df['review_text']]#
```

```python
#same thing for all the sentences: 
s1 = [TextBlob(i).sentences for i in df['review_text']]
```

```python
s1
```

```python
def sentiment(text):
    try:
        return TextBlob(text).sentiment
    except:
        return None
```

```python
df['polarity']=df['review_text'].apply(sentiment).apply(lambda x: x[0])
```

```python
df.head()
```

```python
polrate = df.groupby(['division_name', 'department_name']).polarity.min()
```

```python
polrate
```

```python
polrate_df = pd.DataFrame(polrate)
polrate_df
```

```python
sns.barplot(x = 'division_name', y = 'polarity', data=df)
plt.show()
```

```python
#print out excel and do the rest in Tableau
df.to_excel("women_clothng_sentiment.xlsx", sheet_name='polarity')  
```
